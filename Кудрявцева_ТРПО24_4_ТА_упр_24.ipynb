T остается MST после добавления нового ребра e=(v,w) с весом C, когда C >= максимальному весу ребра на пути между v и w в T. Если c < этого максимума, то T - не MST.


```python
def max_edge(T, v, w):
# T - список смежности, где каждая вершина - список кортежей (соседняя вершина, вес)
    n = len(T)
    parent = [-1] * n
    parent_weight = [0] * n
    visited = [False] * n

    def dfs(u):  # рекурсивный обход в глубину
        visited[u] = True
        if u == w:
            return True
        for nbr, wgt in T[u]:
            if not visited[nbr]:
                parent[nbr] = u
                parent_weight[nbr] = wgt
                if dfs(nbr):
                    return True
        return False

    dfs(v)
    # проходим путь от w до v, ищем максимум
    max_w = -1
    max_edge = None
    cur = w
    while cur != v:
        p = parent[cur]
        wgt = parent_weight[cur]
        if wgt > max_w:
            max_w = wgt
            max_edge = (p, cur)
        cur = p
    return max_w, max_edge

def is_mst(T, v, w, c): # проверяет остаетли ли Т минимальным остовным деревом
    max_w, max_e = max_edge(T, v, w)
    return c >= max_w
```

```python
T = [[(1, 2), (2, 4)],            # вершина 0
      [(0, 2), (3, 3), (4, 5)],   # вершина 1
      [(0, 4), (5, 6)],           # вершина 2
      [(1, 3)],                   # вершина 3
      [(1, 5)],                   # вершина 4
      [(2, 6)]]                   # вершина 5

v, w, c = 3, 5, 4  # Добавляем новое ребро (3,5) с весом 4.
print(f'Ребро с максимальным весом до добавления (v,w): {max_edge(T, v, w)}')
print(f'Т останется минимальным остовным деревом: {is_mst(T, v, w, c)}')
```

```python
import copy
```

```python
def new_mst(T, v, w, c):
# заменяет максимальное ребро на пути v-w на новое ребро (v,w) весом c.

    max_w, (a, b) = max_edge(T, v, w)
    new_T = copy.deepcopy(T) # создаем копию

    new_T[a] = [(x, wgt) for (x, wgt) in new_T[a] if x != b] # удаяляем ребро a-b
    new_T[b] = [(x, wgt) for (x, wgt) in new_T[b] if x != a]

    new_T[v].append((w, c)) # обавляем v-w
    new_T[w].append((v, c))
    return new_T
```

```python
T = [[(1, 2), (2, 4)],            # вершина 0
      [(0, 2), (3, 3), (4, 5)],   # вершина 1
      [(0, 4), (5, 6)],           # вершина 2
      [(1, 3)],                   # вершина 3
      [(1, 5)],                   # вершина 4
      [(2, 6)]]                   # вершина 5


v, w, c = 3, 5, 4  # Добавляем новое ребро (3,5) с весом 4.
print(f'Ребро с максимальным весом до добавления (v,w): {max_edge(T, v, w)}')
print(f'Т останется минимальным остовным деревом: {is_mst(T, v, w, c)}')

new_mst(T, v, w, c)
```
